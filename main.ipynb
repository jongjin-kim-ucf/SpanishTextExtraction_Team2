```python
from DetectExtractText import detect_extract
from TranslateText import translate_text
from SummerizeText import summarize_text

import matplotlib.pyplot as plt

def text_save(text, path):
    import os
    # if not path exist, create it
    if not os.path.exists(os.path.dirname(path)):
        os.makedirs(os.path.dirname(path))  
        
    with open(path, "w") as f:
        f.write(text)
```

```python
for i in range(1, 5):
    img_path = f"training/training{i}.png"
    img_out_path = f"identified_textareas/img{i}.jpg"
    text_out_path = f"identified_text/text{i}.txt"
    translate_text_out_path = f"translated_text/translate_text{i}.txt"
    summary_text_out_path = f"summarized_text/summary_text{i}.txt"


    print(f"############## IMAGE {i} ###############")
    text, img = detect_extract(img_path)

    # print the boxed image
    plt.imsave(img_out_path, img)
    plt.imshow(img)
    plt.show()

    # print and save the detected text
    text_save(text, text_out_path)
    print("Detected Text:")
    print(text)
    print("__________________________\n")

    # translate the text
    print("Translated Text:")
    text_eng = translate_text(text)
    text_save(text_eng, translate_text_out_path)
    print(text_eng)
    print("__________________________\n")

    # summarize the text
    print("Summarized Text:")
    summary = summarize_text(text_eng)
    text_save(summary, summary_text_out_path)
    print(summary)
    print("__________________________\n")
    print("\n\n")
```
